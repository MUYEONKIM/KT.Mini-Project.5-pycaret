```python
train = pd.read_csv('/content/drive/MyDrive/미니프로젝트5차_3_5일차 실습자료/train.csv')
test = pd.read_csv('/content/drive/MyDrive/미니프로젝트5차_3_5일차 실습자료/test_data.csv')

label = test['Activity']

test = test.drop('Activity', axis = 1) #1472 row
```
